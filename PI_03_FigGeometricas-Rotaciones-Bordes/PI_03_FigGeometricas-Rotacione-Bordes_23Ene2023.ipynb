# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

# 1. Figuras geométricas


### Creación de una imagen en color negro

```python
# Imagen color negro de 512 x 512
canvas=np.zeros(shape=(512,512,3),dtype=np.int16 )
canvas.shape
```

```python
plt.imshow(canvas)
plt.show()
```

#### Rectangulo

```python
#cv.rectangle(imagen,origen,final,color,grosor de linea)
cv.rectangle(canvas,pt1=(100,10),pt2=(200,110),color=(0,255,0),thickness=1)
plt.imshow(canvas)
plt.show()
```

```python
cv.rectangle(canvas,pt1=(100,10),pt2=(200,110),color=(0,255,255),thickness=3)
plt.imshow(canvas)
plt.show()
```

```python
cv.rectangle(canvas,pt1=(255,255),pt2=(500,500),color=(0,0,255),thickness=1)
plt.imshow(canvas)
plt.show()
```

#### Circulo

```python
#cv.circle(imagen, centro, radio, color, grosor de linea)
canvas=np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape

cv.circle(canvas,center=(250,250),radius=50,color=(255,255,0),thickness=3)
cv.circle(canvas,center=(250,250),radius=30,color=(0,255,0),thickness=-1)
cv.circle(canvas,center=(250,250),radius=100,color=(255,0,0),thickness=3)

plt.imshow(canvas)
plt.show()
```

```python
plt.imshow(canvas)
plt.show()
```

#### Linea

```python
#cv.line(imagen, inicio, fin, color, grosor de linea)
canvas=np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape

cv.line(canvas,pt1=(0,0),pt2=(250,250),color=(0,255,0),thickness=3)
cv.line(canvas,pt1=(250,250),pt2=(510,0),color=(255,0,0),thickness=3)

plt.imshow(canvas)
plt.show()
```

#### Texto

```python
#font = cv.FONT_HERSHEY_SIMPLEX
#lineType=cv.LINE_AA
#cv.putText(imagen,texto,origen, fuente,tamaño de la fuente,color,grosor, tipo de linea)
canvas=np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape

cv.putText(canvas,text='Hola Mundo',org=(100,200),fontFace=cv.FONT_HERSHEY_SIMPLEX,
           fontScale=1, color=(155,20,105), thickness=3, lineType=cv.LINE_AA)

plt.imshow(canvas)
plt.show()
```


#### Polígono 

```python
#cv.polylines(imagen,puntos,Cerrada o abierta,color y grosor)
canvas=np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape

puntos=np.array(  [[100,200], [400,200], [400,500]])
cv.polylines(canvas, [puntos], isClosed=True, color=(0,255,0), thickness=3)

plt.imshow(canvas)
plt.show()
```

```python
#el mismo poligono pero no cerrado
canvas=np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape

puntos=np.array(  [[100,200], [400,200], [400,500]])
cv.polylines(canvas, [puntos], isClosed=False, color=(0,255,0), thickness=3)

plt.imshow(canvas)
plt.show()
```

## 2.1 Función *flip*

## 3. Detección de bordes

```python
#Visualizar figuras/coins.png
img = cv.imread('figuras/coins.png')
img = cv.cvtColor( img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
img.shape
```

```python
#Pasar a escala de grises
img = cv.imread('figuras/coins.png')
img = cv.cvtColor( img, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.show()
img_gray.shape
```

```python
#Aplicar el filtro GaussianBlur
img = cv.imread('figuras/coins.png')
img = cv.cvtColor( img, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
img_gauss=cv.GaussianBlur(img_gray ,(3,3),0)

plt.imshow(img_gauss,cmap='gray')
plt.show()
img.shape
```

```python
#Aplicar la binarización con la opción cv.THRESH_BINARY_INV
img = cv.imread('figuras/coins.png')
img = cv.cvtColor( img, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
img_gauss=cv.GaussianBlur(img_gray ,(3,3),0)
_, img_thr=cv.threshold(img_gauss, 160,150, cv.THRESH_BINARY_INV)
plt.imshow(img_thr,cmap='gray')
plt.show()
img.shape
```

```python
#Encontrar los contornos con cv.findContours(IMAGEN-BN, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE )
contornos, _=cv.findContours(img_thr, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
```

```python
# Visualizar los contornos  con drawContours
img = cv.imread('figuras/coins.png')
img = cv.cvtColor( img, cv.COLOR_BGR2RGB)
cv.drawContours(img, contornos, -1 , (0,255,0) , 2  )
plt.imshow(img)
plt.show()
img.shape
```
